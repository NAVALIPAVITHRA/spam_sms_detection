IMPORTING THE LIBRARIES

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

# CHANGING THE TEXT INO CSV FILE

```python
import pandas as pd
df = pd.read_csv('/content/spamm.txt', sep=',', encoding='utf-8')
```

```python
df.to_csv('/content/spam.csv', index=False)
```

```python
#DISPLAYS THE FIRST ROWS IN THE DATASET.
df.head(9)
```

```python
data =df.where((pd.notnull(df)),' ')
```

```python
#describing the information in dataset.
df.info()
```

```python
df.shape
```

```python
data.loc[data['v1']== 'spam','v1',]=0
data.loc[data['v1']== 'ham','v1',]=1
```

```python
x=data['v2']
y=data['v1']
```

```python
print(x)
```

```python
print(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
print(x.shape)
print(x_train.shape)
print(x_test.shape)
```

```python
print(y.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
feature_extraction=CountVectorizer(min_df=1,stop_words='english',lowercase=True)
x_feature=feature_extraction.fit_transform(x_train)
test_feature=feature_extraction.transform(x_test)
y_train=y_train.astype('int')
y_test=y_test.astype('int')
```

```python
print(x_feature)
```

# LOGISTIC REGRESSION

```python
model=LogisticRegression()
```

```python
model.fit(x_feature,y_train)
```

```python
prediction_data=model.predict(x_feature)
accuracy_data=accuracy_score(y_train,prediction_data)
```

```python
print('accuracy on training data: ',accuracy_data)
```

```python
prediction_test_data=model.predict(test_feature)
accuracy_test_data=accuracy_score(y_test,prediction_test_data)
```

```python
print('accuracy on testing data: ',accuracy_test_data)
```

# PREDICTING THE DATA

```python
input_email=[" Free entry in 2 a wkly comp to win FA Cup final. "]
input_data=feature_extraction.transform(input_email)
prediction=model.predict(input_data)
print(prediction)
if(prediction[0]==1):
  print("ham mail")
else:
  print("spa mail")
```
